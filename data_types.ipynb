# 5. Binary/Dichotomous Data

```python
Examples:
    1. Heads or Tails
    2. Male or Female
    3. Rich or Poor
    4. Democrate or Republican
    5. BJPian or Congressian 
    6. Pass or Fail
    7. Under age 65 or 65 and over

Binary variables are a sub-type of dichotomous variable, 
"variables assigned either 0 or 1 are said to be in binary state"
Male = 0
Female = 1

Dichotomous ===> Discrete or continous
Dead or alive ---> discrete dichotomous
pass or fail ---> continuous dichotomous
```

```python
example: gender is coded 1 for women and 2 for men
    ----> women are codes as 1 and others as 0
```
