### Exploring DataFrames and Indexes

##### 1. DataFrame

##### is a two dimensional labeled data structure in pandas, representing a rectangular dataset

```python
df1 = pd.DataFrame({'Name': ['Radhe', 'Shiv', 'Ram'], 'Age':[20001, 2302303, 3423433], 'Gender': ['F', 'M', 'M']})
```

```python
df1
```

```python
type(df1)
```

```python
df1.shape
```

##### 2. Series

##### A one-dimensional labeled data structure in pandas, representing a single column or row of a DataFrame

```python
ages = df1["Age"]
```

```python
ages.shape
```

```python
ages1 = pd.Series([25, 43, 29], name="Age")
```

```python
ages1
```

```python
type(ages1)
```

```python
type(ages)
```

##### 3. Index

##### A sequence of labels for the rows or columns of a DataFrame or Series

```python
cust1 = pd.read_csv("./Customers_segmentation.csv", index_col="CustomerID")
```

```python
cust1
```

##### 4. loc

##### A label-based indexing method in pandas for selecting rows and columns by label or boolean mask

```python
df1
```

```python
df1.loc[df1['Age'] >334230, 'Name'] 
```

##### 5. iloc

##### an integer-based indexing method in pandas for selecting rows and columns by integer position or boolean mask

```python
df1.iloc[1:3, :2]  # it is gonna select rows from 1 to 2 and columns from 0 to 1
```

##### 6. GroupBy

##### A function in pandas fro grouping a DataFrame by one or more columns and applying a function to each group

```python
df1.groupby('Gender')['Age'].mean()   # let us calculate the mean age by gender
```

##### 7. Merge

##### A function in pandas for combining two or more DataFrames by a common column or set of columns

```python
dfA = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['narayanan', 'madhava', 'keshava']})
```

```python
dfA
```

```python
dfB = pd.DataFrame({'ID': [2, 3, 4], 'Salary': [453334, 234343, 2434342]})
```

```python
dfB
```

```python
cum = pd.merge(dfA, dfB, on="ID")
```

```python
cum
```

##### 8. pivot table

```python
pd.pivot_table(df1, values='Age', index='Gender', columns='Department', aggfunc="mean" )
```

```python
df1
```

```python
merge vs join : 'merge' is equivalent to SQL's JOIN, 
```

### DataFrames and Indexing 

##### 1. creating a DataFrame

```python
df = pd.DataFrame({'Name': ['Radhe', 'Shiv', 'Ram'], 'Age':[21, 23, 34], 'Gender': ['F', 'M', 'M']})
```

```python
print(df)
```

```python
df.shape
```

##### 2. creating an index

```python
df_I = pd.DataFrame({'Name': ['Radhe', 'Shiv', 'Ram'], 'Age':[21, 23, 34], 'Gender': ['F', 'M', 'M']}, index=['A', 'B', 'C'])
```

```python
df_I
```

##### 3. setting an index 

```python
df_SI = pd.DataFrame({'Name': ['Radhe', 'Shiv', 'Ram'], 'Age':[21, 23, 34], 'Gender': ['F', 'M', 'M']})
```

```python
df_SI
```

```python
df_SI.set_index('Name', inplace=True)
```

```python
print(df_SI)
```

##### 4. resetting an index

```python
df_SI = pd.DataFrame({'Name': ['Radhe', 'Shiv', 'Ram'], 'Age':[21, 23, 34], 'Gender': ['F', 'M', 'M']})
```

```python
df_SI.set_index('Name', inplace=True)
```

```python
print(df_SI)
```

```python
df_SI.reset_index(inplace=True)
```

```python
df_SI
```

##### 5. Hierarchical index

```python
df_H = pd.DataFrame({'Name': ['Radhe', 'Shiv', 'Ram'], 'Age':[21, 23, 34], 'Gender': ['F', 'M', 'M']})
```

```python
df_H
```

```python
df_H.set_index(["Name", "Age"], inplace=True)
```

```python
df_H
```
